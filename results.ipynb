## Tuning R for AutoAction

```python
import numpy as np
import pandas as pd
datasets = ['adult','student','credit100k','credit1m']
```

```python
df = pd.read_csv('./results/exp1.csv')
```

```python
rs = df[["R"]].to_numpy()
rs = rs.reshape((rs.shape[0],))
trials = 5
```

```python
def reader1(dataset, df):
    d = [0 for i in range(df.shape[0])]
    for i in range(trials):
        ds = df[[dataset+str(i+1)]].to_numpy()
        ds = ds.reshape((ds.shape[0],))
        d = d + ds
    return d
```

```python
import matplotlib.pyplot as plt 
def plot1(dataset, df, rs):
    d = reader1(dataset,df)
    d = [abs(s/trials+0.0) for s in d]
    l1=plt.plot(rs,d,'r--')
```

```python
plot1('adult',df,rs)
```

```python
plot1('student',df,rs)
```

```python
plot1('credit100k',df,rs)
```

```python
plot1('credit1m',df,rs)
```

## Supporting constraints

```python
df = pd.read_csv('./results/exp2.csv')
```

```python
rs = df[["time"]].to_numpy()
rs = rs.reshape((rs.shape[0],))

trials = 5
```

```python
def reader2(dataset, df):
    d1 = [0 for i in range(df.shape[0])]
    d2 = [0 for i in range(df.shape[0])]
    for i in range(trials):
        ds1 = df[[dataset+"r"+str(i+1)]].to_numpy()
        ds1 = ds1.reshape((ds1.shape[0],))
        d1 = d1 + ds1
        ds2 = df[[dataset+"a"+str(i+1)]].to_numpy()
        ds2 = ds2.reshape((ds2.shape[0],))
        d2 = d2 + ds2
    
    return d1, d2
```

```python
def plot2(dataset, df, rs):
    d1, d2 = reader2(dataset,df)
    d1 = [abs(s/trials+0.0) for s in d1]
    d2 = [abs(s/trials+0.0) for s in d2]
    l1=plt.plot(rs,d1,'r--')
    l2=plt.plot(rs,d2,'b--')
```

```python
plot2('adult',df,rs)
```

```python
plot2('student',df,rs)
```

```python
plot2('credit100k',df,rs)
```

```python
plot2('credit1m',df,rs)
```
